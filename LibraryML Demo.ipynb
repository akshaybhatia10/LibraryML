Lets look the different components of a Neural Network:

- <b>Input Layer</b>:  The input data(features/labels).
- <b>Arrows</b>:  These are known as weights and biases which the NN modifies to improve itself.
- <b>Hidden Layer</b>: These are essentially a chain or composition of functions (dot product, activation function). The Hidden Layer helps the NN to learn non-linear functions.
- <b>Output Layer</b>:  The output of our network. It can be a continous value(Regression) or a discrete value(Classification).
